Question 9: Interview Logic â€“ Subarray Sum (Hard)

Given a list of integers and a target sum:
nums = [2, 4, 3, 7, 5, 1]
target = 12

Print all pairs of contiguous subarrays whose sum equals the target.

````python
nums = [2, 4, 7, 5, 1]
target = 12

result = []

def find_subsets(index, curr_sum, path):
    if curr_sum == target:
        result.append(tuple(path))
        return
    if curr_sum > target or index == len(nums):
        return

    # include current number
    find_subsets(index + 1, curr_sum + nums[index], path + [nums[index]])

    # exclude current number
    find_subsets(index + 1, curr_sum, path)

find_subsets(0, 0, [])
print(result)
````

Q11. Valid Parentheses

Check if the string is valid.
Input: "()[]{}"
Output: True

Input: "(]"
Output: False

(Use list as stack)

````python
def bracket_checker(s):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for ch in s:
        if ch in pairs.values():       # opening bracket
            stack.append(ch)

        elif ch in pairs:              # closing bracket
            if not stack:
                return False

            if stack[-1] != pairs[ch]:
                return False

            stack.pop()

    return len(stack) == 0


s = input()
print(bracket_checker(s))
````

Q13. Longest Consecutive Sequence
nums = [100, 4, 200, 1, 3, 2]

Output:
4   # sequence: 1,2,3,4

````python
nums = [100, 4, 200, 1, 3, 2]

nums.sort()

max_len = 1
curr_len = 1

for i in range(1, len(nums)):
    if nums[i] == nums[i-1] + 1:
        curr_len += 1
    elif nums[i] != nums[i-1]:
        curr_len = 1   # reset sequence

    if curr_len > max_len:
        max_len = curr_len

print(max_len)
````

Q15. Second Largest (Without Sorting)

Find second largest number from a list.

````python
def second_largest(l):
    largest = float('-inf')
    second_largest = float('-inf')

    for x in l:
        if x > largest:
            second_largest = largest
            largest = x
        elif x > second_largest and x != largest:
            second_largest = x

    return second_largest


list1 = [1, 2, 4, 99, 1821, 2342]
second_largest(list1)
````

Q7. Group Anagrams

Given a list of strings, group anagrams together.

Example:

```
["eat", "tea", "tan", "ate", "nat", "bat"]
```

````python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

anagrams = {}

for word in words:
    key = ''.join(sorted(word))   # step 1: sort characters
    
    if key not in anagrams:
        anagrams[key] = []        # step 2: create group
    
    anagrams[key].append(word)    # step 3: add word

print(list(anagrams.values()))
````

Q10. Rotate Matrix 90Â° Clockwise

Input:
[
 [1, 2, 3],
 [4, 5, 6],
 [7, 8, 9]
]

Output:
[
 [7,4,1],
 [8,5,2],
 [9,6,3]
]

````python
def rotate_matrix(matrix):
    n = len(matrix)

    # Step 1: Transpose
    for i in range(n):
        for j in range(i, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # Step 2: Reverse each row
    for row in matrix:
        row.reverse()

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

rotate_matrix(matrix)
print(matrix)
````

Q15. Recursive Question (Classic)

Find factorial using recursion and explain the call stack.

````python
def fact(n):
    if n <= 1:
        return 1
    return n * fact(n-1)

x = 5
print(fact(x))
````

````python
# Iterative
def fact(n):
    fact = 1
    if n <= 1:
        fact = 1
    else:
        while n > 1:
            fact = fact * n 
            n = n - 1
    return fact
    
x = 5
print(fact(x))
````
