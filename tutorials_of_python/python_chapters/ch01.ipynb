In Python, everything is an <span style="color:red"><em>*object*</em></span> which means <span style="color:red"><em>*None*</em></span>(which means variable has no values or not pointing to any object yet) is also an object!

<span style="color:red"><em>*is*</em></span> checks if two variables are referring to the same objects
<span style="color:red"><em>==</em></span> checks if two variables points to have same values

```python
a = [1, 2, 3]
b = a
b += [4]
print(a)

# print(b = a) #TypeError: print() got an unexpected keyword argument 'b'
print(b==a)
print(b is a)

b = [1, 2, 3, 4]
print(b is a) # False when a and b don't refers to the same object
print(b==a) # True when a and b pointing to are equal(the same value of an object)
```

Local vs Global Variables

```python
age = 18 # global variable

def isEligible(user_age):
    if user_age >= age: # user_age is local variable
        return 'Great! You can vote'
    else:
        return 'Wait for next year!'
    
print(isEligible(5))
print(isEligible(25))
```

Types of variables

Type casting: conversion of variables from one data-type to another
Implicit type casting vs Explicit type casting

```python
a = "123"
print(a)
print(type(a))
b = int(a) # Explicit type casting - assign the converted value to another variable
print(b)
print(type(b))
print(float(a)) # Implicit type casting - inline conversion
```

Operators in Python

Arithmetic operators: -,+,*, **, /, //, %

Logical operators: and, or, not

Comparision operators: ==, !=, <, >, <=, >=

Bitwise Shift (left, right): >>, <<

Bitwise: &, |, ~, ^

Assignment operators: =, +=, -=, *=, %=

```python
a, b = 5, 4
print(a+b, a-b, a*b, a**b, a/b, a//b, a%b) # arithmetic
print(a < b, a > b, a>=b, a<=b, a==b, a!=b) # comparison 
x, y = 1, 0
print(x and y, x or y, not x) # logical
print(x << 1, y >> 1) # bitwise shift
print(x&y, x|y, ~y, x^y, ~x^y) # bitwise
print(1^1)
```

```python
a = 5
b = 2

# Identity Operators - is, is not
print(a is b)
print(a is not b)

# Membership Operators - in, not in
vowels = ['a', 'e', 'i', 'o', 'u']
if 'a' in vowels:
    print('Found')
elif 'a' not in vowels:
    print('Not Found')

# Tenary Operators - conditional expressions are operators that evaluate something based on a condition being true or false
min = a if a < b else b

print(min)
```
