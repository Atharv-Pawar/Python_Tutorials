In Python, everything is an <span style="color:red"><em>*object*</em></span> which means <span style="color:red"><em>*None*</em></span>(which means variable has no values or not pointing to any object yet) is also an object!

<span style="color:red"><em>*is*</em></span> checks if two variables are referring to the same objects
<span style="color:red"><em>==</em></span> checks if two variables points to have same values

```python
a = [1, 2, 3]
b = a
b += [4]
print(a)

# print(b = a) #TypeError: print() got an unexpected keyword argument 'b'
print(b==a)
print(b is a)

b = [1, 2, 3, 4]
print(b is a) # False when a and b don't refers to the same object
print(b==a) # True when a and b pointing to are equal(the same value of an object)
```

Local vs Global Variables

```python
age = 18 # global variable

def isEligible(user_age):
    if user_age >= age: # user_age is local variable
        return 'Great! You can vote'
    else:
        return 'Wait for next year!'
    
print(isEligible(5))
print(isEligible(25))
```

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)


arr = [3, 6, 11, 5, 1, 2, 1]
print(quicksort(arr))
```
