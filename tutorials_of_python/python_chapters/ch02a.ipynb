Tips & Tricks

```python
s = 'a'
s2 = 'A'
"""When you compare two strings, Python compares them character by character using their Unicode code points (similar to ASCII for common English letters)."""
print(ord('a'))  # 97
print(ord('A'))  # 65
print(s < s2)  # comparison of strings
s = 'abc'
s2 = 'ab'
print(s > s2)  # comparison of strings
s = 'zzz'
s2 = 'aaaaaa'
print(s > s2)  # comparison of strings
# swapping of variables
a = 5
b = 7
print("a, b", a, b)
temp = a
a = b
b = temp
print("a, b", a, b)
# shortcut
a, b = 3, 4
print("a, b", a, b)
a, b = b, a
print("a, b", a, b)
```

**Conditional Statement**

<span style="color:Red">if-else, if-elif-else, nested if</span>

```python
# if-else
age = int(input("Enter your age: "))

if age >= 17:
    print("You can issue a driver's license.")
else:
    print("Wait! You are not old enough.")
```

```python
# if-elif-else
age = int(input("Enter your age: "))

if age >= 18:
    print("You can issue a driver's license.")
elif age >= 16:
    print("You can apply for a learner's permit.")
else:
    print("Wait! You are not old enough.")
```

```python
# nested if
age = int(input("Enter your age: "))
state = input("Enter your state: ")

if state.lower() == "california":
    if age >= 18:
        print("You can apply for a driver's license.")
    elif age >= 16:
        print("You can apply for a learner's permit.")
    else:
        print("Wait! You are not old enough.")
elif state.lower() == "texas":
    if age >= 18:
        print("You can apply for a driver's license.")
    elif age >= 15:
        print("You can apply for a learner's permit.")
    else:
        print("Wait! You are not old enough.")
elif state.lower() == "new york":
    if age >= 17:
        print("You can apply for a driver's license.")
    elif age >= 15:
        print("You can apply for a learner's permit.")
    else:
        print("Wait! You are not old enough.")
else:
    print("Invalid response")
```

**Looping**

<span style="color:Red">for</span>

```python
for i in range(100):
    # do something
    pass
```

```python
for i in range(5):
    print("The value of i is currently", i)

for k in range(2, 8, 3):
    print("The value of k is currently", k)
```

<span style="color:Red">break & continue</span>

```python
for i in range(1, 10):
    if i == 5:
        break
    else:
        print(f"Now the square of {i} is", i**2)
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    else:
        print(f"The value of i is currently {i}")
```

<span style="color:Red">while</span>

```python
count = 5 # count = 0
while count != 0: # while count != 5
    print(count)
    count -= 1 # count += 1
```

```python
# infinite loop
# run the below to check the infinite loop output
while True:
    print("I'm stuck inside a loop.")
```
