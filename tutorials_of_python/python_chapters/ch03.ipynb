<span style="color:Red">*args</span> Arbitrary Arguments: The **\*args** parameter allows a function to accept any number of positional arguments.<br>
<span style="color:Red">**kwargs</span> Arbitrary Keyword Arguments: The **\*\*kwargs** parameter allows a function to accept any number of keyword arguments.

```python
def coffee_order(customer_name, coffee_type):
    print(customer_name, "ordered a", coffee_type, "coffee.")

coffee_order("Alice", "Latte")
# coffee_order("Bob", "Espresso", "Cappuccino") # TypeError: coffee_order() takes 2 positional arguments but 3 were given
```

```python
# *args is the tuple of all extra parameters 
# tuple named "args"
def coffee_order(customer_name, coffee_type, *other_items): # here *other_items is *args i.e. a tuple containing all extra positional arguments
    print(customer_name, "ordered a", coffee_type, "coffee.")
    if other_items:
        print("Additional items ordered by", customer_name, "are:")
        for item in other_items:
            print("-", item)

coffee_order("Alice", "Latte")
coffee_order("Bob", "Espresso", "Cappuccino", "Mocha", "Americano", "Cold Brew")
```

```python
# **kwargs is a dictionary containing all extra keyword arguments
# dictionary named "kwargs"
def coffee_order(customer_name, coffee_type, **kwargs):
    print(customer_name, "ordered a", coffee_type, "coffee.")
    if kwargs:
        print("Additional details for", customer_name, "are:")
        for key, value in kwargs.items():
            print("-", key, ":", value)

coffee_order("Samira", "Espresso", size="large", syrup="vanilla")
coffee_order("Carolina", "Cappuccino", size="small", syrup="caramel")
```

```python
# while using *args and **kwargs together in a function definition, *args must be written first and then **kwargs
def coffee_order(customer_name, coffee_type, *args, **kwargs):
    print(customer_name, "ordered a", coffee_type, "coffee.")
    if args:
        print("Additional items ordered by", customer_name, "are:")
        for item in args:
            print("-", item)
    if kwargs:
        print(f"Additional details for {customer_name}'s coffee order are:")
        for key, value in kwargs.items():
            print("-", key, ":", value)

coffee_order("Joe", "Americano", "Croissant", "Muffin", size="medium", syrup="caramel", temperature="hot")
coffee_order("Zara", "Cold Coffee", size="large", syrup="honey")
```

```python
def tea_order(customer_name, tea_type, *args, **kwargs):
    print(customer_name, "ordered a", tea_type, "tea.")
    if args:
        print("Additional items ordered by", customer_name, "are:")
        for item in args:
            print("-", item)
    if kwargs:
        print(f"Additional details for {customer_name}'s tea order are:")
        for key, value in kwargs.items():
            print("-", key, ":", value)

emma_extras = ("Scone", "Muffin") # emma is regular customer and her regular order includes these items
harry_extras = {"size": "small", "syrup": "caramel", "flavor": "lemon-ginger"} # harry has specific preferences for his tea order

tea_order("Emma", "Green Tea", *emma_extras)
tea_order("Harry", "Black Tea", **harry_extras)
```

```python
def average_scores(*scores):
    if len(scores) == 0:
        return 0
    total = sum(scores)
    return total / len(scores)

print(average_scores(85, 90, 78, 92, 88))
print(average_scores(81,79,93))
print(average_scores())
```

```python
def make_sandwich(*ingredients, **options):
    print("Making a sandwich with the following ingredients:")
    for ingredient in ingredients:
        print("-", ingredient)
    if options:
        print("Additional options:")
        for key, value in options.items():
            print("-", key, ":", value)

make_sandwich("turkey", "cheese", "lettuce", "tomato", bread="whole wheat", sauce="mayo", toasted=True)
```

Decorators <span style="color:Red">@</span>

Decorators let you add extra behavior to a function, without changing the function's code.
A decorator is a function that takes another function as input and returns a new function.

```python
import time

def brew_tea():
    print("Brewing tea...")
    time.sleep(2)  # Simulate brewing time
    print("Tea is ready!")

brew_tea()
```

```python
import time

def brew_tea():
    start_time = time.time()
    print("Brewing tea...")
    time.sleep(2)  # Simulate brewing time
    print("Tea is ready!")
    end_time = time.time()
    print(f"Tea brewing time: {end_time - start_time:.6f} seconds")

brew_tea()
```

here the function brew_tea() violates the rules, it is focusing on two tasks messaging "Tea is ready" as well as calculating execution time duration.

a function must focus on one single responsibility or single task

```python
def brew_tea():
    start_time = time.time()
    print("Brewing tea...")
    time.sleep(2)  # Simulate brewing time
    print("Tea is ready!")
    end_time = time.time()
    print(f"Tea brewing time: {end_time - start_time:.6f} seconds")

def make_matcha():
    start_time = time.time()
    print("Making matcha...")
    time.sleep(1)  # Simulate matcha making time
    print("Matcha is ready!")
    end_time = time.time()
    print(f"Matcha making time: {end_time - start_time:.6f} seconds")

brew_tea()
make_matcha()
```

here also, we have to calculate execution time and repeating code in a program is not a good practice 

Hence, decorator comes into picture.

```python
# first method to apply decorator
def timer_dec(base_func):
    def enhance_func():
        start_time = time.time()
        base_func()
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
    return enhance_func

def brew_tea():
    print("Brewing tea...")
    time.sleep(2)  # Simulate time taken to brew tea
    print("Tea is ready!")

dec_brew_tea = timer_dec(brew_tea)
dec_brew_tea()
brew_tea() # considering brew_tea and dec_brew_tea are different

brew_tea = timer_dec(brew_tea)
brew_tea() # considering brew_tea with time of execution functionality 
```

```python
# second method to apply decorators
def timer_dec(base_func):
    def enhance_func():
        start_time = time.time()
        base_func()
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
    return enhance_func

@timer_dec
def brew_tea():
    print("Brewing tea...")
    time.sleep(2)
    print("Tea is ready!")

brew_tea()
```

```python
def timer_dec(base_func):
    def enhance_func(tea_type, sleep_time):
        start_time = time.time()
        base_func(tea_type, sleep_time)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
    return enhance_func

@timer_dec
def brew_tea(tea_type, sleep_time):
    print(f"Brewing {tea_type}...")
    time.sleep(sleep_time)
    print(f"{tea_type} is ready!")

@timer_dec
def make_matcha():
    print("Making matcha...")
    time.sleep(1)
    print("Matcha is ready!")

brew_tea("Green Tea", 1)
# make_matcha() # TypeError because we need to pass two arguments (tea_type, sleep_time) as written in enhanced_func to fix this we will used *args, **kwargs
```

```python
def timer_dec(base_func):
    def enhance_func(*args, **kwargs):
        start_time = time.time()
        base_func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
    return enhance_func

@timer_dec
def brew_tea(tea_type, sleep_time):
    print(f"Brewing {tea_type}...")
    time.sleep(sleep_time)
    print(f"{tea_type} is ready!")

@timer_dec
def make_matcha():
    print("Making matcha...")
    time.sleep(1)
    print("Matcha is ready!")

brew_tea("Green Tea", 1)
brew_tea(tea_type="Black Tea", sleep_time=2)
make_matcha()
```

```python
import time
from datetime import datetime, timedelta

def timer_dec(base_func):
    def enhance_func(*args, **kwargs):
        start_time = time.time()
        result = base_func(*args, **kwargs) # decorator will capture the return value
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
        return result # decorator will return the result of the original function
    return enhance_func

@timer_dec
def brew_tea(tea_type, sleep_time):
    print(f"Brewing {tea_type}...")
    time.sleep(sleep_time)
    print(f"{tea_type} is ready!")

@timer_dec
def make_matcha():
    print("Making matcha...")
    time.sleep(1)
    print("Matcha is ready!")
    return f"Drink matcha by {datetime.now() + timedelta(minutes=5)}" # Return the scheduled time

brew_tea(tea_type="Black Tea", sleep_time=2)
print(make_matcha())
```

**Generators**

<span style="color:Red">yield</span>

```python
def generate_values():
    yield 1
    yield 2
    yield 3

g = generate_values()
print(next(g))
print(next(g))
print(next(g))
print(next(g)) # StopIteration
```

```python
def func(number):
    yield number ** 2

for i in range(1, 6):
    print(next(func(i)))
```

```python
def gen_primes(start, end):
    for i in range(start, end + 1):
        if i < 2:
            continue
        is_prime = True

        for j in range(2, int(i ** 0.5) + 1):
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            yield i

gen_primes_obj = gen_primes(50, 120)
print(next(gen_primes_obj)) # GENERATORS = lazy evaluation, consumes less memory, makes code more modular & reusable, 
# On other hand the function gives complete result at once is called = eager evaluation
```

Lambda Function

```python
# Lambda function: it can have multiple parameters, must be written in single line, must have one expression, anonymous function as they don't have any name
lambda num: num ** 2
```

```python
def square(num):
    return num ** 2

def cube(num):
    return num ** 3

def transform_list(num_list, transform_item):
    return [transform_item(num) for num in num_list]

my_list = [2, 5, 3]
print(transform_list(my_list, square))
print(transform_list(my_list, cube))
```

```python
def transform_num(num_list, transform_function):
    return [transform_function(num) for num in num_list]

transform_num([2, 7, 3], lambda x: x ** 2)
```

```python
# map() function
num_list = [2, 7, 3, 5, 8, 6]

squared_list = list(map(lambda x: x**2, num_list))
print(squared_list)

# filter() function
even_list = list(filter(lambda num: num %2 == 0, num_list))
print(even_list)
```

```python
# Use the sorted() function and a lambda to sort the list by score in descending order.

students = [('Alice', 88), ('Bob', 72), ('Charlie', 95)]
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)
print(sorted_students)
```
