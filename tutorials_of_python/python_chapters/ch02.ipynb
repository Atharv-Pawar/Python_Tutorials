**The Enumerate Function**

The *enumerate()* function allows you to loop through a sequence (like a list, tuple, or string) while keeping track of the index position of each item.

without *enumerate()* we would access both the index and the value the following way:

```python
fruits = ["apple", "banana", "orange"]
for i in range(len(fruits)):
    print(f"Index {i}: {fruits[i]}")
```

*enumerate()* is a more elegant way to get both index and value:

```python
fruits = ["apple", "banana", "orange"]
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")
```

<span style="color:Green"><strong>**Binary**</strong></span>

In Python, bytes, bytearray, and memoryview are binary sequence types used to work with raw binary data efficiently.
They differ mainly in mutability, memory usage, and performance.

<span style="color:Red">bytes</span><br>
Immutable sequence of bytes (0–255).<br>
Similar to a string, but stores raw bytes instead of Unicode characters.<br>
Often used for read-only binary data (e.g., file contents, network packets).

```python
# Creating a bytes object
data = bytes([65, 66, 67]) # ASCII values for 'A', 'B', 'C'
print(data) # Output: b'ABC'
# Accessing elements
print(data[0]) # Output: 65
# Slicing
print(data[1:]) # Output: b'BC'
```

Immutable: Cannot be modified after creation.<br>
Supports slicing, indexing, and iteration.<br>
Useful for representing binary data like files or network packets.

<span style="color:Red">bytearray</span><br>
The bytearray type is a mutable counterpart of bytes. It allows modification of its contents and is ideal for scenarios requiring frequent updates to binary data.

```python
# Creating a bytearray
data = bytearray([65, 66, 67]) # ASCII values for 'A', 'B', 'C'
print(data) # Output: bytearray(b'ABC')
# Modifying elements
data[0] = 97 # ASCII for 'a'
print(data) # Output: bytearray(b'aBC')
# Appending and deleting
data.append(68) # ASCII for 'D'
print(data) # Output: bytearray(b'aBCD')
del data[1]
print(data) # Output: bytearray(b'aCD')
```

Mutable: Can be modified in place. <br>
Supports list-like operations such as append, insert, and del.<br>
Useful for performance-critical applications like image or real-time data processing.

<span style="color:Red">memoryview</span><br>
The memoryview type provides a way to access the memory of an object supporting the buffer protocol (e.g., bytes, bytearray) without copying the data. This is particularly useful for handling large datasets efficiently.

```python
# Creating a memoryview
data = bytearray(b'Hello')
view = memoryview(data)
# Accessing and modifying data
print(view[0]) # Output: 72 (ASCII for 'H')
view[1] = 105 # ASCII for 'i'
print(data) # Output: bytearray(b'Hillo')
# Slicing
sub_view = view[1:4]
print(sub_view.tobytes()) # Output: b'ill'
```

Avoids copying data, improving performance for large datasets.<br>
Allows slicing and indexing.<br>
Can reinterpret data using the .cast() method to change its type

<span style="color:Red">bytes:</span> Reading binary files, sending immutable data over a network.<br>
<span style="color:Red">bytearray:</span> Modifying binary data, such as image or audio processing.<br>
<span style="color:Red">memoryview:</span> Efficiently handling large binary datasets without copying.

**Stack Implementation**

A stack follows the **Last-In-First-Out (LIFO)** principle. You can use Python's list or collections.deque to implement it.

```python
# Stack implementation using list
stack = []
# Push elements onto the stack
stack.append("A")
stack.append("B")
stack.append("C")
print("Stack after pushes:", stack)
# Pop elements from the stack
print("Popped element:", stack.pop())
print("Stack after pop:", stack)
```

**Queue Implementation**

A queue follows the **First-In-First-Out (FIFO)** principle. While list can be used, collections.deque is more efficient for queues.

```python
from collections import deque
# Queue implementation using deque
queue = deque()
# Enqueue elements
queue.append("X")
queue.append("Y")
queue.append("Z")
print("Queue after enqueues:", queue)
# Dequeue elements
print("Dequeued element:", queue.popleft())
print("Queue after dequeue:", queue)
```

<span style="color:Red">Stack Operations:</span> append() for push. pop() for pop.<br>
<span style="color:Red">Queue Operations:</span> append() for enqueue. popleft() (from deque) for dequeue.
